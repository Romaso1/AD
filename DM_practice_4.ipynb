#### **Система Лоренца:**

Система Лоренца описується трьома диференціальними рівняннями:

$$
\begin{cases}
\frac{dx}{dt} = \sigma (y - x) \\
\frac{dy}{dt} = x (\rho - z) - y \\
\frac{dz}{dt} = xy - \beta z
\end{cases}
$$

де:
- $( \sigma )$ — коефіцієнт Прандтля (зазвичай $ \sigma = 10 $),
- $( \rho )$ — числовий параметр для турбулентності (зазвичай $ \rho = 28 $),
- $( \beta )$ — коефіцієнт відношення висоти до довжини (зазвичай $ \beta = 8/3 $).

#### **Об'єкти Нарендра-Пархтізаратхі:**

Об'єкти Нарендра-Пархтізаратхі описуються наступними рівняннями:

$$
\begin{cases}
x_1(t+1) = \alpha_1 x_1(t) + \beta_1 x_2(t) + \gamma_1 x_3(t) + \delta_1 \\
x_2(t+1) = \alpha_2 x_1(t) + \beta_2 x_2(t) + \gamma_2 x_3(t) + \delta_2 \\
x_3(t+1) = \alpha_3 x_1(t) + \beta_3 x_2(t) + \gamma_3 x_3(t) + \delta_3
\end{cases}
$$

де $(\alpha_i)$, $(\beta_i)$, $(\gamma_i)$ і $(\delta_i)$ — коефіцієнти для кожного з рівнянь.

### Параметри:
- $( a_1, a_2 )$, $( b_1, b_2 )$ — це коефіцієнти, які можуть бути варійовані для дослідження динаміки системи та впливу на прогнозування.

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

#система лоренца
def lorenz_system(t, state, sigma=10.0, rho=28.0, beta=8.0/3.0):
    x, y, z = state
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]
```

```python
#модель Нарендра
def narendra_parthasarathy(t, state, alpha1=0.5, alpha2=0.4, beta1=0.6, beta2=0.2):
    x1, x2 = state
    dx1dt = alpha1 * x1 + beta1 * np.tanh(x2)
    dx2dt = alpha2 * x2 + beta2 * np.tanh(x1)
    return [dx1dt, dx2dt]
```

```python
#ген данних для системи лоренца
def generate_lorenz_data(t_span, t_eval, initial_state):
    solution = solve_ivp(lorenz_system, t_span, initial_state, t_eval=t_eval)
    return solution.y.T  # повернув матрицю
```

```python
#ген данних нарендра
def generate_np_data(t_span, t_eval, initial_state):
    solution = solve_ivp(narendra_parthasarathy, t_span, initial_state, t_eval=t_eval)
    return solution.y.T  # повернув матрицю
```

```python
#прогноз значень
def predict_with_linear_regression(data, future_steps):
    X = data[:-1]
    y = data[1:]
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
    
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    predictions = model.predict(X_test)
    mse = mean_squared_error(y_test, predictions)
    
    #прогноз майбутніх знач
    future_predictions = []
    last_value = data[-1]
    for _ in range(future_steps):
        next_value = model.predict(last_value.reshape(1, -1))
        future_predictions.append(next_value[0])
        last_value = next_value
    
    return predictions, future_predictions, mse
```

```python
#результати
def plot_results(true_data, predicted_data, future_predictions, title):
    plt.plot(true_data, label='True Data')
    plt.plot(range(len(predicted_data)), predicted_data, label='Predicted Data', linestyle='--')
    plt.plot(range(len(true_data), len(true_data) + len(future_predictions)), future_predictions, label='Future Predictions', linestyle=':')
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
if __name__ == "__main__":
    # ген лоренц
    t_span = (0.0, 50.0)
    t_eval = np.linspace(0.0, 50.0, 1000)
    initial_state_lorenz = [1.0, 1.0, 1.0]
    lorenz_data = generate_lorenz_data(t_span, t_eval, initial_state_lorenz)

    # ген нарендр
    initial_state_np = [0.1, 0.1]
    np_data = generate_np_data(t_span, t_eval, initial_state_np)

    # прогноз лоренц
    predictions_lorenz, future_predictions_lorenz, mse_lorenz = predict_with_linear_regression(lorenz_data, future_steps=50)
    print(f"Mean Squared Error for Lorenz System: {mse_lorenz}")
    plot_results(lorenz_data[:, 0], predictions_lorenz[:, 0], future_predictions_lorenz, "Lorenz System Predictions")

    # прогноз нарендр
    predictions_np, future_predictions_np, mse_np = predict_with_linear_regression(np_data, future_steps=50)
    print(f"Mean Squared Error for Narendra-Parthasarathy Model: {mse_np}")
    plot_results(np_data[:, 0], predictions_np[:, 0], future_predictions_np, "Narendra-Parthasarathy Model Predictions")
```
